```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
crimeratedata = pd.read_csv('final_crimes.csv')
crimeratedata.head()
```

```python
nj = crimeratedata['state_abbr']=='NJ'
```

```python
njcrimerate = crimeratedata.loc[nj]
njcrimerate['auto_theft_rate'] = (njcrimerate.motor_vehicle_theft/njcrimerate.population)*100
njcrimerate
```

```python
# Assign the data to X and y
# Note: Sklearn requires a two-dimensional array of values
# so we use reshape to create this

X = njcrimerate.year.values.reshape(-1, 1)
y = njcrimerate.auto_theft_rate.values.reshape(-1, 1)

print("Shape: ", X.shape, y.shape)
```

```python
plt.scatter(X, y,marker="o", edgecolors="black",s=y*200, alpha=0.75,c='blue')
plt.grid()
#clb = plt.colorbar()
plt.xlabel("YEAR")
plt.ylabel("MOTOR VEHICLE THEFT RATE ")
plt.gcf().set_size_inches(10,8)
#plt.savefig('nj_carTheft_scatterplot.png')
```

```python

from sklearn.model_selection import train_test_split

### BEGIN SOLUTION

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

### END SOLUTION
```

```python
# Create the model

### BEGIN SOLUTION

from sklearn.linear_model import LinearRegression

model = LinearRegression()

### END SOLUTION
```

```python
# Fit the model to the training data. 

### BEGIN SOLUTION

model.fit(X_train, y_train)

### END SOLUTION
```

```python
# Calculate the mean_squared_error and the r-squared value
# for the testing data

from sklearn.metrics import mean_squared_error, r2_score

### BEGIN SOLUTION

# Use our model to make predictions
predicted = model.predict(X_test)

# Score the predictions with mse and r2
mse = mean_squared_error(y_test, predicted)
r2 = r2_score(y_test, predicted)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2 ): {r2}")
```

```python
# Call the `score` method on the model to show the r2 score

### BEGIN SOLUTION
model.score(X_test, y_test)
### END SOLUTION
```

```python
# Fit the model to the training data and calculate the scores for the training and testing data

### BEGIN SOLUTION
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

### END SOLUTION 

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
# Plot the Residuals for the Training and Testing data

### BEGIN SOLUTION   marker="o", edgecolors="red",s=y/500, alpha=0.75,c=y
plt.scatter(model.predict(X_train), model.predict(X_train) - y_train, c="blue", label="Training Data",marker="o")
plt.scatter(model.predict(X_test), model.predict(X_test) - y_test, c="orange", label="Testing Data")
plt.legend()
plt.grid()
plt.hlines(y=0, xmin=y.min(), xmax=y.max())
plt.title("NJ Car Theft Residual Plot")
plt.xlabel('X')
plt.ylabel('Residual Values')
plt.gcf().set_size_inches(10,8)
plt.savefig('nj_carTheft_residualplot.png')
```

```python
# Note: we have to transform our min and max values 
# so they are in the format: array([[ 1.17]])
# This is the required format for `model.predict()`

x_min = np.array([[X.min()]])
x_max = np.array([[X.max()]])
print(f"Min X Value: {x_min}")
print(f"Max X Value: {x_max}")
```

```python
# Calculate the y_min and y_max using model.predict and x_min and x_max

### BEGIN SOLUTION
y_min = model.predict(x_min)
y_max = model.predict(x_max)
### END SOLUTION
```

```python
# Plot X and y using plt.scatter
# Plot the model fit line using [x_min[0], x_max[0]], [y_min[0], y_max[0]]

### BEGIN SOLUTION
plt.scatter(X, y, c='blue')
#clb = plt.colorbar()
plt.plot([x_min[0], x_max[0]], [y_min[0], y_max[0]], c='red')
plt.title("NJ Auto Theft Rate Trend")
plt.xlabel("Year")
plt.ylabel("Car Theft")
plt.grid(True)
plt.gcf().set_size_inches(10,8)
plt.legend(('Predicted Car Theft', 'Actual Car Theft'), loc='best', framealpha=0.25, prop={'size':'small', 'family':'monospace'})
plt.savefig('CarTheft_trend.png')
### END SOLUTION
```
