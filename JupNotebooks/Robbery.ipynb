```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import numpy
```

```python
crimeratedata = pd.read_csv('final_crimes.csv')
crimeratedata.head()
```

```python
nj = crimeratedata.loc[crimeratedata['state_abbr'] == 'NJ']
nj
```

```python
nj["robbery_rate"] = nj["robbery"] / nj["population"]*100
nj
```

```python
X = nj.year.values.reshape(-1, 1)
y = nj.robbery_rate.values.reshape(-1, 1)

print("Shape: ", X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

### BEGIN SOLUTION

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
### BEGIN SOLUTION
plt.scatter(model.predict(X_train), model.predict(X_train) - y_train, c="blue", label="Training Data")
plt.scatter(model.predict(X_test), model.predict(X_test) - y_test, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y.min(), xmax=y.max())
plt.title("NJ Robbery Rate Residual Plot")
plt.xlabel('Independent Variable: X')
plt.ylabel('Residual Values')
plt.savefig('nj_robbery_residualplot.png')
### END SOLUTION
```

```python
x_min = np.array([[X.min()]])
x_max = np.array([[X.max()]])
y_min = model.predict(x_min)
y_max = model.predict(x_max)
```

```python
plt.scatter(X, y, c='blue')
plt.plot([x_min[0], x_max[0]], [y_min[0], y_max[0]], c='red')
plt.title("NJ Robbery/Population Rate Trend")
plt.xlabel("Year")
plt.ylabel("% of Robbery by Population")
plt.grid(True)
plt.gcf().set_size_inches(8,4)
plt.legend(('Predicted Robbery', 'Actual Robbery'), loc='best', framealpha=0.25, prop={'size':'small', 'family':'monospace'})
plt.savefig('robbery_trend.png')
```
