```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import numpy
```

```python
crimeratedata = pd.read_csv('final_crimes.csv')
crimeratedata.head()
```

```python
nj = crimeratedata.loc[crimeratedata['state_abbr'] == 'NJ']
nj
```

```python
nj["homicide_rate"] = nj["homicide"] / nj["population"]*100
nj
```

```python

X = nj.year.values.reshape(-1, 1)
y = nj.homicide_rate.values.reshape(-1, 1)

print("Shape: ", X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

### BEGIN SOLUTION

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python

x_min = np.array([[X.min()]])
x_max = np.array([[X.max()]])
y_min = model.predict(x_min)
y_max = model.predict(x_max)
```

```python
plt.scatter(X, y, c='blue')
plt.plot([x_min[0], x_max[0]], [y_min[0], y_max[0]], c='red')
plt.title("NJ Homcide/Population Rate Trend")
plt.xlabel("Year")
plt.ylabel("% of Homicide by Population")
plt.grid(True)
plt.gcf().set_size_inches(8,4)
plt.legend(('Predicted Homicide', 'Actual Homicide'), loc='best', framealpha=0.25, prop={'size':'small', 'family':'monospace'})
plt.savefig('homicide_trend.png')
```
